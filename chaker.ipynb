```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Split the dataset into features (X) and target variables (y)
X = food_df.drop(columns=['GI', 'GL'])
y_gi = food_df['GI']
y_gl = food_df['GL']

```

```python
# Split the dataset into training and testing sets (80% train, 20% test)
X_train, X_test, y_train_gi, y_test_gi = train_test_split(X, y_gi, test_size=0.2, random_state=42)
X_train, X_test, y_train_gl, y_test_gl = train_test_split(X, y_gl, test_size=0.2, random_state=42)
```

```python
# Initialize and train linear regression models for GI and GL
gi_model = LinearRegression()
gi_model.fit(X_train, y_train_gi)

gl_model = LinearRegression()
gl_model.fit(X_train, y_train_gl)
```

```python
# Predict GI and GL on the test set
y_pred_gi = gi_model.predict(X_test)
y_pred_gl = gl_model.predict(X_test)
```

```python
# Evaluate the performance of the models
gi_rmse = mean_squared_error(y_test_gi, y_pred_gi, squared=False)
gi_r2 = r2_score(y_test_gi, y_pred_gi)

gl_rmse = mean_squared_error(y_test_gl, y_pred_gl, squared=False)
gl_r2 = r2_score(y_test_gl, y_pred_gl)

print("GI Model Performance:")
print("RMSE:", gi_rmse)
print("R-squared:", gi_r2)

print("\nGL Model Performance:")
print("RMSE:", gl_rmse)
print("R-squared:", gl_r2)
```

Based on these results, it seems that the regression model for predicting GL performs significantly better than the one for predicting GI. This may be due to the stronger correlations between the independent variables and GL compared to GI.

```python
# Retrieve the coefficients for GI model
gi_coefficients = gi_model.coef_
gi_attributes = X.columns

# Retrieve the coefficients for GL model
gl_coefficients = gl_model.coef_
gl_attributes = X.columns

# Create DataFrame to display coefficients for GI model
gi_coefficients_df = pd.DataFrame({'Attribute': gi_attributes, 'Coefficient': gi_coefficients})
gi_coefficients_df = gi_coefficients_df.sort_values(by='Coefficient', ascending=False)
print("Coefficients for GI Model:")
print(gi_coefficients_df)

# Create DataFrame to display coefficients for GL model
gl_coefficients_df = pd.DataFrame({'Attribute': gl_attributes, 'Coefficient': gl_coefficients})
gl_coefficients_df = gl_coefficients_df.sort_values(by='Coefficient', ascending=False)
print("\nCoefficients for GL Model:")
print(gl_coefficients_df)
```

```python
import matplotlib.pyplot as plt

# Plot feature importance for GI model
plt.figure(figsize=(10, 6))
plt.barh(gi_coefficients_df['Attribute'], gi_coefficients_df['Coefficient'], color='skyblue')
plt.xlabel('Coefficient')
plt.ylabel('Attribute')
plt.title('Feature Importance for GI Model')
plt.gca().invert_yaxis()  # Invert y-axis to display attribute with the highest importance at the top
plt.show()

# Plot feature importance for GL model
plt.figure(figsize=(10, 6))
plt.barh(gl_coefficients_df['Attribute'], gl_coefficients_df['Coefficient'], color='salmon')
plt.xlabel('Coefficient')
plt.ylabel('Attribute')
plt.title('Feature Importance for GL Model')
plt.gca().invert_yaxis()  # Invert y-axis to display attribute with the highest importance at the top
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize and train Random Forest regressor for GI and GL
rf_gi_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_gi_model.fit(X_train, y_train_gi)

rf_gl_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_gl_model.fit(X_train, y_train_gl)

rf_y_pred_gi = rf_gi_model.predict(X_test)
rf_y_pred_gl = rf_gl_model.predict(X_test)

rf_gi_rmse = mean_squared_error(y_test_gi, rf_y_pred_gi, squared=False)
rf_gi_r2 = r2_score(y_test_gi, rf_y_pred_gi)

rf_gl_rmse = mean_squared_error(y_test_gl, rf_y_pred_gl, squared=False)
rf_gl_r2 = r2_score(y_test_gl, rf_y_pred_gl)


print("\nRandom Forest Model Performance:")
print("GI - RMSE:", rf_gi_rmse, " R-squared:", rf_gi_r2)
print("GL - RMSE:", rf_gl_rmse, " R-squared:", rf_gl_r2)
```

```python
# Compute residuals for GI and GL predictions
gi_residuals = y_test_gi - rf_y_pred_gi
gl_residuals = y_test_gl - rf_y_pred_gl

# Calculate mean and variance of residuals for GI and GL
gi_residual_mean = np.mean(gi_residuals)
gi_residual_variance = np.var(gi_residuals)

gl_residual_mean = np.mean(gl_residuals)
gl_residual_variance = np.var(gl_residuals)

print("Mean and Variance of Residuals for GI Model:")
print("Mean:", gi_residual_mean)
print("Variance:", gi_residual_variance)

print("\nMean and Variance of Residuals for GL Model:")
print("Mean:", gl_residual_mean)
print("Variance:", gl_residual_variance)
```

## Classification 

```python
# Function to classify GI values
def classify_gi(value):
    if value < 55:
        return 'Low'
    elif 55 <= value < 70:
        return 'Medium'
    else:
        return 'High'

# Function to classify GL values
def classify_gl(value):
    if value <= 10:
        return 'Low'
    elif 11 <= value <= 19:
        return 'Medium'
    else:
        return 'High'

```

```python
# Apply classification to GI and GL values
food_df['GI_Class'] = food_df['GI'].apply(classify_gi)
food_df['GL_Class'] = food_df['GL'].apply(classify_gl)

# Display the first few rows with the classified GI and GL values
print(food_df[['GI', 'GI_Class', 'GL', 'GL_Class']].head())
```

```python
# Plot the distribution of GI categories
plt.figure(figsize=(8, 6))
sns.countplot(x='GI_Class', data=food_df, order=['Low', 'Medium', 'High'])
plt.title('Distribution of GI Categories')
plt.xlabel('GI Category')
plt.ylabel('Count')
plt.show()

# Plot the distribution of GL categories
plt.figure(figsize=(8, 6))
sns.countplot(x='GL_Class', data=food_df, order=['Low', 'Medium', 'High'])
plt.title('Distribution of GL Categories')
plt.xlabel('GL Category')
plt.ylabel('Count')
plt.show()
```

```python
# Select relevant features (attributes) for classification
X_classification = food_df[['Protein', 'Carb', 'DF', 'TF', 'CAL', 'FF']]

# Target variables for classification
y_gi_classification = food_df['GI_Class']
y_gl_classification = food_df['GL_Class']
```

```python
# Split the dataset into training and testing sets for GI classification
X_train_gi, X_test_gi, y_train_gi, y_test_gi = train_test_split(X_classification, y_gi_classification, test_size=0.2, random_state=42)

# Split the dataset into training and testing sets for GL classification
X_train_gl, X_test_gl, y_train_gl, y_test_gl = train_test_split(X_classification, y_gl_classification, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report

# Initialize and train Decision Tree classifier for GI classification
dt_classifier_gi = DecisionTreeClassifier(random_state=42)
dt_classifier_gi.fit(X_train_gi, y_train_gi)

# Initialize and train Decision Tree classifier for GL classification
dt_classifier_gl = DecisionTreeClassifier(random_state=42)
dt_classifier_gl.fit(X_train_gl, y_train_gl)

# Predictions for GI classification
dt_y_pred_gi = dt_classifier_gi.predict(X_test_gi)

# Predictions for GL classification
dt_y_pred_gl = dt_classifier_gl.predict(X_test_gl)

# Evaluate Decision Tree classifier for GI classification
print("Classification Report for GI (Decision Tree Classifier):")
print(classification_report(y_test_gi, dt_y_pred_gi))

# Evaluate Decision Tree classifier for GL classification
print("\nClassification Report for GL (Decision Tree Classifier):")
print(classification_report(y_test_gl, dt_y_pred_gl))
```
